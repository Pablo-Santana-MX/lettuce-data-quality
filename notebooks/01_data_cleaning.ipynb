<p style="text-align: center;">
This notebook documents the data cleaning and preparation process for lettuce growth analysis.<br>
It includes: column renaming, missing value handling, duplicate removal, date normalization,<br>
derived variable creation, and numeric scaling for analysis and predictive models.
</p>


```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# 1️⃣ Cargar dataset CSV
df = pd.read_csv("https://raw.githubusercontent.com/Pablo-Santana-MX/lettuce-data-quality/main/data/raw/lettuce_growth_raw.csv", encoding="latin1")

print("Columnas originales:", df.columns.tolist())

# 2️⃣ Renombrar directamente a nombres estándar en inglés
df.rename(columns={
    "Plant_ID": "plant_id",
    "Date": "date",
    "Temperature (°C)": "temperature_c",
    "Humidity (%)": "humidity_percent",
    "TDS Value (ppm)": "tds_ppm",
    "pH Level": "ph_level",
    "Growth Days": "growth_days"
}, inplace=True)

print("Columnas renombradas:", df.columns.tolist())

# 3️⃣ Manejo de valores nulos
print("Valores nulos por columna:\n", df.isnull().sum())
df['ph_level'] = df['ph_level'].fillna(df['ph_level'].mean())
df = df.dropna(subset=['temperature_c', 'humidity_percent'])

# 4️⃣ Eliminar duplicados
df = df.drop_duplicates()

# 5️⃣ Normalizar fechas
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# 6️⃣ Conversión de tipos
df['plant_id'] = df['plant_id'].astype(int)
df['growth_days'] = df['growth_days'].astype(int)

# 7️⃣ Crear variables derivadas
def clasificar_humedad(h):
    if h < 40:
        return "Low"
    elif 40 <= h <= 70:
        return "Medium"
    else:
        return "High"

df['humidity_class'] = df['humidity_percent'].apply(clasificar_humedad)

def clasificar_temp(t):
    if t < 25:
        return "Low"
    elif 25 <= t <= 30:
        return "Optimal"
    else:
        return "High"

df['temperature_class'] = df['temperature_c'].apply(clasificar_temp)

# 8️⃣ Escalado numérico (para modelos)
scaler = MinMaxScaler()
df[['temperature_c','humidity_percent','tds_ppm','ph_level']] = scaler.fit_transform(
    df[['temperature_c','humidity_percent','tds_ppm','ph_level']]
)

# ✅ Validación final
print(df.head())
print(df['humidity_class'].value_counts())
print(df['temperature_class'].value_counts())

```
